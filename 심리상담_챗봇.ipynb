```python
import pandas as pd
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity # chatbot 대화 유사도 계산하기 위해
```

```python
model = SentenceTransformer('jhgan/ko-sroberta-multitask')

sentences = ["안녕하세요?", "한국어 문장 임베딩을 위한 버트 모델입니다"]
```

```python
embeddings = model.encode(sentences)

print(embeddings)
```

```python
df = pd.read_excel('/content/웰니스_대화_스크립트_데이터셋.xlsx')
df.head()
```

```python
# 챗봇에 있는 NaN은 필요없는 데이터니 삭제
df = df[~df['챗봇'].isna()]
df.head()
```

```python
# model.encode(df.loc[0, '유저'])
```

```python
df['embedding'] = pd.Series([[]] * len(df))
df['embedding'] = df['유저'].map(lambda x: list(model.encode(x)))
df.head()
```

```python
df.to_csv('wellness_dataset.csv', index=False) # 위의 데이터를 해당 이름.csv으로 저장
```

```python
text = '요즘 머리가 아프고 너무 힘들어'
embedding = model.encode(text)
```

```python
df['distance'] = df['embedding'].map(lambda x:cosine_similarity([embedding], [x]). squeeze())
df.head()
```

```python
answer = df.loc[df['distance'].idxmax()]
```

```python
print('구분: ', answer['구분'])
print('유사한 질문: ', answer['유저'])
print('챗봇 답변: ', answer['챗봇'])
print('유사도: ', answer['distance'])
```
